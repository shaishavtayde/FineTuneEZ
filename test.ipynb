```python
import pandas as pd

df = pd.read_csv('train.csv')
df = df.rename(columns={'sms':'text'})

df

```

```python
def split_datasets(df):
    train = 0.7
    val = 0.15
    test = 0.15

    train_size = int(len(df) * train)
    val_size = int(len(df) * val)
    test_size = int(len(df) * test)
    
    train_df = df.iloc[:train_size]
    val_df = df.iloc[train_size:train_size + val_size]
    test_df = df.iloc[train_size + val_size:]
    
    return train_df, val_df, test_df
```

```python
train_d, val_d, test_d = split_datasets(df[:100])
train_d

```

```python
from datasets import Dataset, DatasetDict

train_dataset = Dataset.from_pandas(train_d)

val_dataset = Dataset.from_pandas(val_d)

test_dataset = Dataset.from_pandas(test_d)

dataset_dict = DatasetDict({
    "train": train_dataset,
    "valid": val_dataset,
    "test": test_dataset
})

```

```python
import os
# Save the dataset in Arrow format
base_dir = "dataset"

dataset_dict.save_to_disk(base_dir)

print(f"Dataset saved successfully in .arrow format at {base_dir}")
```
